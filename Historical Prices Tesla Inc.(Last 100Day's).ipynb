```python
# Importing Required Library
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Set the path to the ChromeDriver executable
chrome_driver_path = r"C:\Users\Swapnil Alaspure\Documents\MIT Exam 2nd\WebScraping\chromedriver.exe"
# Create ChromeOptions object
chrome_options = webdriver.ChromeOptions()
# Instantiate the Chrome WebDriver with the specified options
driver = webdriver.Chrome(options=chrome_options)
# Open the URL
driver.get("https://finance.yahoo.com/quote/TSLA/history?period1=1293840000&period2=1700179200&interval=1d&filter=history&frequency=1d&includeAdjustedClose=true")
```

```python
# Our list empty list
table_headers = []
name_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/thead/tr'
name = driver.find_element(By.XPATH, name_xpath).text
table_headers.append(name)
# Display the list
table_headers
```

```python
# Rearranging 
tableHeaders = ['Date' ,'Open', 'High', 'Low Close*', 'Adj Close**' ,'Volume']
tableHeaders
```

```python
# Our list empty list
players_data = []
name_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody'
name = driver.find_element(By.XPATH, name_xpath).text
players_data.append(name.replace(' ', ','))
# Display the list
players_data
```

```python
# Date

Date = []

for i in range(1,101):
    
    Date_xpath= '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i) +']/td[1]'
    
    date = driver.find_element(By.XPATH,
                                 Date_xpath).text
    
    Date.append(date)
    
Date
```

```python
# Open

Open = []

for i in range(1,101):
    
    Open_xpath='//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i) +']/td[2]'
    
    open = driver.find_element(By.XPATH,
                                 Open_xpath).text
    
    Open.append(open)
    
Open
```

```python
# High

High = []

for i in range(1,101):
    
    High_xpath= '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i)+']/td[3]'
    
    high = driver.find_element(By.XPATH,
                                 High_xpath).text
    
    High.append(high)
    
High
```

```python
# Low

Low = []

for i in range(1,101):
    
    Low_xpath= '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i) +']/td[4]'
    
    low = driver.find_element(By.XPATH,
                                 Low_xpath).text
    
    Low.append(low)
    
Low
```

```python
# Close*

Close = []

for i in range(1,101):
    
    Close_xpath= '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i) +']/td[5]'
    
    close = driver.find_element(By.XPATH,
                                 Close_xpath).text
    
    Close.append(close)
    
Close
```

```python
# Adj Close**

Adj_Close = []

for i in range(1,101):
    
    AdjClose_xpath= '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i) +']/td[6]'
    
    adj = driver.find_element(By.XPATH,
                                 AdjClose_xpath).text
    
    Adj_Close.append(adj)
    
Adj_Close
```

```python
# Volume

Volume = []

for i in range(1,101):
    
    Volume_xpath= '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+ str(i) +']/td[7]'
    
    Vlm = driver.find_element(By.XPATH,
                                 Volume_xpath).text
    
    Volume.append(Vlm)
    
Volume
```

```python
# Converting Into Data Frame
Historical_Prices_Tesla = pd.DataFrame({ "Date":Date,
                                         "Open":Open,
                                         "High":High,
                                         "Low":Low,
                                         "Close*":Close,
                                         "Adj Clsoe**": Adj_Close,
                                         "Volume":Volume})

# Display Data Frame
Historical_Prices_Tesla
```

```python
# Close the Driver
driver.close()
```
